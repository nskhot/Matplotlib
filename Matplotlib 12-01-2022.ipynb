# Pyplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# Markers

```python
## HOW TO USE MARKERS
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'or')
plt.show()
```

# How to use Default X-Points

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

#plt.plot(ypoints, marker = 'o')
plt.plot(ypoints, marker = "d")
plt.show()
```

```python
#You can also use the shortcut string notation parameter 
# to specify the marker.

#This parameter is also called fmt (format), and is written with 
# this syntax:"marker|line|color"


ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

# plt.plot(ypoints, '+:b')
# plt.plot(ypoints, 's-.y')
plt.plot(ypoints, 'D--m')
plt.show()
```

# Marker Size

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 15)
plt.show()

```

# Marker Color

```python
# mec : marker edge color

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o--r', ms = 20, mec = 'b')
plt.show()
```

```python
# mfc: marker face color
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'k')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o-.', ms = 20, mec = 'b', mfc = 'y')
plt.show()
```

# Labels

```python
## Explode a wedge

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

# Shadow

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = [0.2, 0, 0, 0]


plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True)
plt.show() 

```

# Legend

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = [0.2, 0, 0, 0]


plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True)

plt.legend(title = "Four Fruits:", loc = 2)
plt.show() 

```

```python
# LEgend with header
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:",loc=4)
plt.show()
```
